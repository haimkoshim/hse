## Задание №1

Скачайте текст "Литературных анекдотов". Напишите функцию, которая будет читать файл, лемматизировать текст с помощью pymystem3 и записывать результат в новый файл. У функции должно бы два аргумента: путь к исходному файлу и путь к файлу с лемматизированным текстом. Вызов функции тоже должен быть прописан в решении.

```python
from pymystem3 import Mystem
import re


m = Mystem()
```

```python
def lemmatization(dir_input, dir_output):
    with open(dir_input, 'r') as f:
        read_data = f.read()
    read_data = read_data.replace("\n", " ")
    read_data = re.sub(r'[^\w\s]', '', read_data).lower()
    
    lemmas = m.lemmatize(read_data)
   
    with open(dir_output, 'w+') as g:
        g.write(''.join(lemmas))
        
#   output
    with open(dir_output, 'r') as g:
        data = g.read()
    print(data)
    

lemmatization(r'C:\Users\Данил\Desktop\task_1.txt', r'C:\Users\Данил\Desktop\task_1_1.txt')
```

## Задание №2

Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.

```python
import nltk
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

sw = stopwords.words('russian')

with open(r'C:\Users\Данил\Desktop\task_1_1.txt', 'r') as g:
    data = g.read()
data = data.split()

filtered = [word for word in data if word not in sw]
print(filtered)
```

```python
from collections import Counter

counts = Counter(filtered)

#ipm
for word in list(counts):
    print(word + ':' + ' ipm = ' + str(filtered.count(word)/sum(counts.values())*1000000))
```

```python
#top 20
print(counts.most_common(20))
```

## Задание №3

Сделайте полный морфологический разбор исходного текста. Напишите регулярное выражение, которое будет извлекать из тега только часть речи. Пройдитесь циклом по списку с разборами, который выдал pymystem3, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение). Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
with open(r'C:\Users\Polina\Desktop\task_1.txt', 'r') as g:
    data = g.read()
m.analyze(data)
```

```python
read_data = data.replace("\n", " ")
read_data = re.sub(r'[^\w\s]', '', read_data).lower()

res = m.analyze(read_data)

d = dict()

part_of_speech = []

for i in range (len(res)): 
    try:
        grammar = res[i]['analysis'][0]['gr']

        result = re.findall(r'^\w+', grammar)
        word = res[i]['text']
        
        array = []
        array.append(word)
        array.append(''.join(result))
        part_of_speech.append(''.join(result))
        
        d.update([array])
        
    except (KeyError, IndexError):
        pass

print(d)
```

```python
co = Counter(part_of_speech)

print('absolute_numers :', '\n')
for item in co:
    print(item, co[item])
print('\n')

print('relative_numbers :', '\n')
for item in co:
    print(item, co[item]/sum(co.values())*1000000)
```
